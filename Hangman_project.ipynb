```python
import random
# Intro
print("Welcome to hangman - guess the major subject in mathematics !")
print("""
 _                                                                 _   _
| |                                                               | | | |
| |__   __ _ _ __   __ _ _ __ ___   __ _ _ __      _ __ ___   __ _| |_| |__
| '_ ' / _` | '_ ' / _` | '_ ` _ ' / _` | '_ '    | '_ ` _ ' / _` | __| '_ '
| | | | (_| | | | | (_| | | | | | | (_| | | | |   | | | | | | (_| | |_| | | |
|_| |_|'__,_|_| |_|'__, |_| |_| |_|'__,_|_| |_|   |_| |_| |_|'__,_|'__|_| |_|
                    __/ |
                   |___/
""")


# 난이도 선택: 단어의 길이로 분류된 난이도 레벨 선택
x = input("Choose difficulty (Easy/ Medium/ Hard): ")


# 도전자가 오타 등으로 잘못 입력했으면 다시 입력 요청
if x.lower() not in ["easy", "medium", "hard"]:
  print("Choose in easy, medium, hard")
  x = input("Choose difficulty (Easy/ Medium/ Hard): ")

# 선택한 난이도의 단어 리스트에서 단어 하나를 랜덤하게 추출
if x.lower() == "easy":
  word = ["analysis", "topology", "set theory", "statistics"]
  a = random.choice(word)
  hint = 0 # 난이도별로 사용할 수 있는 힌트의 개수: easy/ medium/ hard 각 난이도 별로 0개/ 1개/ 2개 사용 가능

elif x.lower() == "medium":
  word = ["probability", "linear algebra", "combinatorics", "field theory", "number theory", "algebraic topology", "abstract algebra"]
  a = random.choice(word)
  hint = 1

elif x.lower() == "hard":
  word = ["differential geometry", "numerical analysis", "mathematical logic", "discrete mathematics", "partial differential equation", "ordinary differential equation"]
  a = random.choice(word)
  hint = 2


# 도전자에게 단어의 글자 수 알려주기: 알파벳 자리는 "■", 빈칸(띄어쓰기) 자리는 "_"로 표기
l = []
for i in range(0, len(a)):
  if a[i] == " ":
    l.append("_")
  else:
    l.append("■")
print(l)

# 처음 행맨 그림 출력
print(drawings[0])

# 도전자가 알파벳을 하나씩 입력하며 단어 맞추기 - 입력한 알파벳이 단어에 있다면 그 알파벳 자리에서 "■" 대신 알파벳을 넣어 출력
                                                # 알파벳이 단어에 없다면 행맨 아스키 아트 출력
remaining = 6
l2 = [] # 도전자가 시도한 알파벳 전체를 모으는 리스트
l3 = [] # 도전자가 시도한 알파벳 중 단어에 없는 알파벳을 모으는 리스트

while remaining >= 0:
  # 도전자에게 힌트를 사용할 건지 질문
  if hint > 0: # 사용할 수 있는 힌트가 남아 있는지 확인
    x = input("Would you like to use a hint? (Yes/ No): ")
    x = x.lower()
    if x != "yes" and x != "no":
      x = input("Would you like to use a hint? (Yes/ No): ")
      x = x.lower()
    elif x == "yes": # 도전자가 힌트를 사용할 경우 단어의 알파벳 중 도전자가 입력하지 않은 알파벳 하나를 랜덤하게 골라 종류와 위치를 알려주기
      l4 = []
      for i in range(0, len(a)):
        if i != " " and i not in l:
          l4.append(a[i])
      y = random.choice(l4)
      for i in range(0, len(a)):
        if a[i] == y:
          l[i] = y

      print(l)
      hint = hint - 1

      if hint > 0: # 사용할 수 있는 힌트가 남아 있으면 같은 과정을 한번 더 진행
        x = input("Would you like to use a hint? (Yes/ No): ")
        x = x.lower()
        if x != "yes" and x != "no":
          x = input("Would you like to use a hint? (Yes/ No): ")
          x = x.lower()
        elif x == "yes":
          l4 = []
          for i in range(0, len(a)):
            if i != " " and i not in l:
              l4.append(a[i])
          y = random.choice(l4)
          for i in range(0, len(a)):
            if a[i] == y:
              l[i] = y

          print(l)
          hint = hint - 1



  # 도전자가 알파벳을 하나씩 입력하며 단어 맞추기 - 입력한 알파벳이 단어에 있다면 그 알파벳 자리에서 "■" 대신 알파벳을 넣어 출력
                                                # 알파벳이 단어에 없다면 행맨 아스키 아트 출력
  x = input("your guess alphabet: ")
  x = x.lower()

  # 도전자가 실수로 알파벳이 아닌 문자를 입력했으면 다시 입력 요청
  if x not in "abcdefghijklmnopqrstuvwxyz":
    print("You must input an alphabet!")
    x = input("your guess alphabet: ")

  # 도전자가 입력한 알파벳이 이미 이전에 시도한 알파벳이면 다시 입력 요청
  if x in l2:
    print(f"You already tried {x}!")
    x = input("your guess alphabet: ")

  l2.append(x)
  if x in a:
    for i in range(0, len(a)):
      if a[i] == x:
        l[i] = x

  if x not in a:
    print(f"{x} is not in the word!")
    l3.append(x)
    print(drawings[l3.index(x)+1])
    remaining = remaining - 1

  print(l)

  # 도전자가 알파벳을 모두 맞추어 단어가 완성되었으면 축하의 메세지 출력
  if "■" not in l:
    print("Congratulations! you win the game!")
    print("""
                                                                                         88                    88
                                                                       ,d                88              ,d    ""
                                                                       88                88              88
 ,adPPYba,  ,adPPYba,  8b,dPPYba,   ,adPPYb,d8 8b,dPPYba, ,adPPYYba, MM88MMM 88       88 88 ,adPPYYba, MM88MMM 88  ,adPPYba,  8b,dPPYba,  ,adPPYba,
a8"     "" a8"     "8a 88P'   `"8a a8"    `Y88 88P'   "Y8 ""     `Y8   88    88       88 88 ""     `Y8   88    88 a8"     "8a 88P'   `"8a I8[    ""
8b         8b       d8 88       88 8b       88 88         ,adPPPPP88   88    88       88 88 ,adPPPPP88   88    88 8b       d8 88       88  `"Y8ba,
"8a,   ,aa "8a,   ,a8" 88       88 "8a,   ,d88 88         88,    ,88   88,   "8a,   ,a88 88 88,    ,88   88,   88 "8a,   ,a8" 88       88 aa    ]8I
 `"Ybbd8"'  `"YbbdP"'  88       88  `"YbbdP"Y8 88         `"8bbdP"Y8   "Y888  `"YbbdP'Y8 88 `"8bbdP"Y8   "Y888 88  `"YbbdP"'  88       88 `"YbbdP"'
                                    aa,    ,88
                                     "Y8bbdP"         """)
    break

  # 도전자가 6번의 기회를 모두 사용해 게임이 끝났으면 행맨 그림이 완성되고 "Game over" 메세지 출력
  if remaining == 0:
    print("You lose the game..")
    print(f"The word was {a}!")
    print("""⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀
                                   ⠀           ⠀⣀⣀⠀⠀⢀⣤⣤⣤⣶⣶⣷⣤⣀
⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⢀⣶⣶⣶⠀⠀⠀⠀⣠⣾⣿⣿⡇⠀⣿⣿⣿⣿⠿⠛⠉⠉⠀
⠀⠀⠀⠀⠀⠀⠀⠀⠀⣀⣀⣀⣀⠀⠀⠀⠀⠀⢀⣿⣿⣶⡀⠀⠀⠀⠀⠀⣾⣿⣿⣿⡄⠀⢀⣴⣿⣿⣿⣿⠁⢸⣿⣿⣿⣀⣤⡀⠀⠀⠀
⠀⠀⠀⠀⠀⣠⣴⣶⣿⣿⣿⣿⣿⣷⠀⠀⠀⠀⣼⣿⣿⣿⣧⠀⠀⠀⠀⢰⣿⣿⣿⣿⣇⣠⣿⣿⣿⣿⣿⡏⢠⣿⣿⣿⣿⣿⡿⠗⠂⠀⠀
⠀⠀⠀⣰⣾⣿⣿⠟⠛⠉⠉⠉⠉⠋⠀⠀⠀⣰⣿⣿⣿⣿⣿⣇⣠⣤⣤⣿⣿⣿⢿⣿⣿⣿⣿⢿⣿⣿⡿⠀⣼⣿⣿⡟⠉⠁⢀⣀⡄⠀⠀
⠀⢀⣾⣿⡿⠋⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⢰⣿⣿⣿⣴⣿⣿⣿⣿⡿⣿⣿⣿⡏⠈⢿⣿⣿⠏⣾⣿⣿⠃⢠⣿⣿⣿⣶⣶⣿⣿⣿⡷⠦⠀
⢠⣾⣿⡿⠀⠀⠀⣀⣠⣴⣶⣿⣿⡷⠀⣠⣿⣿⣿⣿⡿⠟⣿⣿⣿⣠⣿⣿⣿⠀⠀⠀⠀⠁⢸⣿⣿⡏⠀⣼⣿⣿⣿⠿⠛⠛⠉⠀⠀⠀⠀
⢸⣿⣿⠣⣴⣾⣿⣿⣿⣿⣿⣿⡿⠃⣰⣿⣿⣿⠋⠁⠀⠀⠸⣿⣿⣿⣿⣿⠃⠀⠀⠀⠀⠀⠸⠿⠿⠀⠀⠛⠛⠟⠀⠀⠀⠀⠀⠀⠀⠀⠀
⠸⣿⣿⣆⣉⣻⣭⣿⣿⣿⡿⠋⠀⠀⢿⣿⡿⠁⠀⠀⠀⠀⠀⠹⠟⠛⠛⠛⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀
⠀⠙⠿⣿⣿⣿⣿⡿⠟⠋⠀⠀⠀⠀⠀⠈⠁⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀
⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⣀⣤⣤⣶⣶⣶⣶⣦⣄⠀⠀
⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⢀⣷⠄⣤⣤⣤⣤⣶⣾⣷⣴⣿⣿⣿⣿⠿⠿⠛⣻⣿⣿⣷⡄
⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⣀⣠⣤⣄⠀⣶⣶⣤⡀⠀⠀⠀⠀⠀⠀⢀⣴⣿⠋⢠⣿⣿⣿⠿⠛⠋⠉⠛⣿⣿⣿⠏⢀⣤⣾⣿⣿⡿⠋⠀
⠀⠀⠀⠀⠀⠀⠀⠀⣠⣴⣾⣿⣿⣿⣿⠓⢹⣿⣿⣷⠀⠀⠀⠀⢀⣶⣿⡿⠁⠀⣾⣿⣿⣟⣠⣤⠀⠀⢸⣿⣿⣿⣾⣿⣿⣿⡟⠋⠀⠀⠀
⠀⠀⠀⠀⠀⠀⣠⣾⣿⣿⡿⠛⠉⠸⣿⣦⡈⣿⣿⣿⡇⠀⠀⣰⣿⣿⡿⠁⠀⢸⣿⣿⣿⣿⣿⠿⠷⢀⣿⣿⣿⣿⡿⠛⣿⣿⣿⡀⠀⠀⠀
⠀⠀⠀⠀⢀⣼⣿⣿⡿⠋⠀⠀⠀⠀⣿⣿⣧⠘⣿⣿⣿⡀⣼⣿⣿⡟⠀⠀⢀⣿⣿⣿⠋⠁⠀⣀⣀⣼⣿⣿⡟⠁⠀⠀⠘⣿⣿⣧⠀⠀⠀
⠀⠀⠀⠀⣼⣿⣿⡟⠀⠀⠀⠀⠀⣠⣿⣿⣿⠀⢹⣿⣿⣿⣿⣿⡟⠀⠀⠀⣼⣿⣿⣷⣶⣿⣿⣿⣿⣿⣿⡟⠀⠀⠀⠀⠀⠸⣿⣿⡆⠀⠀
⠀⠀⠀⠀⢹⣿⣿⣇⠀⠀⢀⣠⣴⣿⣿⣿⡿⠀⠈⣿⣿⣿⣿⡟⠀⠀⠀⢰⣿⣿⣿⠿⠟⠛⠉⠁⠸⢿⡟⠀⠀⠀⠀⠀⠀⠀⠘⠋⠁⠀⠀
⠀⠀⠀⠀⠈⢻⣿⣿⣿⣾⣿⣿⣿⣿⣿⠟⠁⠀⠀⠸⣿⣿⡿⠁⠀⠀⠀⠈⠙⠛⠃⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀
⠀⠀⠀⠀⠀⠀⠉⠛⠿⠿⠿⠿⠟⠋⠁⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀""")
    break

```
